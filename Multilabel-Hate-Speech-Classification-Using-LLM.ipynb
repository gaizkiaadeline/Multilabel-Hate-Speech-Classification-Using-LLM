```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import re
import string
import warnings

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report

from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from transformers import DataCollatorWithPadding
from datasets import Dataset, DatasetDict
import torch.nn as nn

warnings.filterwarnings('ignore')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
data = pd.read_csv('data_1B.csv')
data
```

```python
data['labels'] = data[['HS', 'Abusive', 'HS_Individual', 'HS_Group', 'HS_Religion', 'HS_Race', 'HS_Physical', 'HS_Gender', 'HS_Other']].values.tolist()

# Tokenize and preprocess datset
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
def preprocess_function(examples):
    return tokenizer(examples['Tweet'], truncation=True, padding='max_length', max_length=128)

# Convert to Hugging Face dataset
dataset = Dataset.from_pandas(data)
encoded_dataset = dataset.map(preprocess_function, batched=True)

# Split dataset
train_dataset, temp_dataset = encoded_dataset.train_test_split(test_size=0.2, shuffle=True, seed=42).values()
val_dataset, test_dataset = temp_dataset.train_test_split(test_size=0.5, shuffle=True, seed=42).values()

# Create DatasetDict for Trainer
dataset_dict = DatasetDict({
    'train': train_dataset,
    'validation': val_dataset,
    'testing': test_dataset
})

```

Bert Tokenizer (Bidirectional Encoder Representations from Transformers) berfungsi untuk memproses teks menjadi representasi numerik yang dapat dimengerti oleh model BERT.


Bert Tokenizer menggunakan teknik tokenisasi WordPiece, yang membagi kata-kata menjadi potongan-potongan yang lebih kecil dan dapat mengatasi out-of-vocabulary (OOV) dengan cara mempunyai nilai dimensi yang dapat di atur.



Keuntungan dari memakai model BERT sendiri adalah dapat merepresentasi bahasa lebih baik, mampu memahami konteks yang lebih luas, fleksibel, dan dapat disesuaikan.

```python
# Define model and custom loss function for multilabel classification
class MultilabelBERT(BertForSequenceClassification):
    def __init__(self, config):
        super().__init__(config)
        self.num_labels = config.num_labels
        self.classifier = nn.Linear(config.hidden_size, config.num_labels)

    def forward(self, input_ids=None, attention_mask=None, token_type_ids=None, position_ids=None, head_mask=None, inputs_embeds=None, labels=None, output_attentions=None, output_hidden_states=None, return_dict=None):
        outputs = self.bert(
            input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids,
            position_ids=position_ids,
            head_mask=head_mask,
            inputs_embeds=inputs_embeds,
            output_attentions=output_attentions,
            output_hidden_states=output_hidden_states,
            return_dict=return_dict,
        )

        pooled_output = outputs[1]
        logits = self.classifier(pooled_output)

        loss = None
        if labels is not None:
            loss_fct = nn.BCEWithLogitsLoss()
            loss = loss_fct(logits, labels.float())
            return {'loss': loss, 'logits': logits}

        return {'logits': logits}

# Initialize model
model = MultilabelBERT.from_pretrained('bert-base-uncased', num_labels=9)
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

# Training arguments
training_args = TrainingArguments(
    output_dir='./results',
    eval_strategy='epoch',
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
)

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = (logits > 0).astype(int)
    f1 = f1_score(labels, predictions, average='weighted')
    acc = accuracy_score(labels, predictions)
    return {'f1': f1, 'accuracy': acc}

# Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset_dict['train'],
    eval_dataset=dataset_dict['validation'],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

# Train and evaluate
trainer.train()
results = trainer.evaluate()
print(results)
```

```python
# Evaluate on validation set
val_results = trainer.evaluate(eval_dataset=dataset_dict['validation'])
print("Validation Results:", val_results)

# Evaluate on test set
test_results = trainer.evaluate(eval_dataset=dataset_dict['testing'])
print("Test Results:", test_results)
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = (logits > 0).astype(int)
    f1 = f1_score(labels, predictions, average='weighted')
    precision = precision_score(labels, predictions, average='weighted')
    recall = recall_score(labels, predictions, average='weighted')
    accuracy = accuracy_score(labels, predictions)
    return {'f1': f1, 'precision': precision, 'recall': recall, 'accuracy': accuracy}

def train_and_evaluate(learning_rate, batch_size):
    training_args = TrainingArguments(
        output_dir='./results',
        eval_strategy='epoch',
        learning_rate=learning_rate,
        per_device_train_batch_size=batch_size,
        per_device_eval_batch_size=batch_size,
        num_train_epochs=3,
        weight_decay=0.01,
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=dataset_dict['train'],
        eval_dataset=dataset_dict['validation'],
        tokenizer=tokenizer,
        data_collator=data_collator,
        compute_metrics=compute_metrics
    )

    trainer.train()
    results = trainer.evaluate()
    return results
```

```python
# Hyperparameter tuning
best_f1 = 0
best_params = {}

for lr in [3e-5, 5e-5]:
    for bs in [16, 32]:
        results = train_and_evaluate(lr, bs)
        if results['eval_f1'] > best_f1:
            best_f1 = results['eval_f1']
            best_params = {'learning_rate': lr, 'batch_size': bs}

print(f"Best F1 Score: {best_f1}")
print(f"Best Hyperparameters: {best_params}")
```

```python
# Train final model with best hyperparameters
final_results = train_and_evaluate(best_params['learning_rate'], best_params['batch_size'])
```

```python
# Evaluate on test set
trainer = Trainer(
    model=model,
    args=TrainingArguments(
        output_dir='./results',
        eval_strategy='epoch',
        per_device_train_batch_size=best_params['batch_size'],
        per_device_eval_batch_size=best_params['batch_size'],
        num_train_epochs=3,
        weight_decay=0.01,
    ),
    train_dataset=dataset_dict['train'],
    eval_dataset=dataset_dict['testing'],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)
```

```python
test_results = trainer.evaluate(dataset_dict['testing'])

print("Final Training Results:", final_results)
print("Test Results:", test_results)
```




*   eval_loss: Nilai loss yang dihasilkan pada saat evaluasi pada data. Loss yang lebih rendah menunjukkan model yang lebih baik dalam memprediksi data yang diamati.
*   eval_f1: Skor F1 adalah harmonic mean dari precision dan recall, digunakan untuk mengukur akurasi dari suatu model klasifikasi.
*   eval_precision: Precision mengukur seberapa banyak dari prediksi positif yang sebenarnya positif.
*   eval_recall: Recall mengukur seberapa banyak dari keseluruhan kelas positif yang diprediksi dengan benar oleh model.
*   eval_accuracy: Akurasi dari evaluasi pada data. Akurasi mengukur seberapa banyak prediksi yang benar dari total prediksi.
*   eval_runtime: Waktu yang diperlukan untuk melakukan evaluasi pada data, dalam detik.
*   eval_samples_per_second: Jumlah sampel yang dievaluasi per detik pada data.
*   eval_steps_per_second: Jumlah langkah evaluasi yang diselesaikan per detik pada data.
